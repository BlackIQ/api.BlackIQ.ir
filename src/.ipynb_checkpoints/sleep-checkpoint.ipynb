```python
import matplotlib.pyplot as plt
import requests as req
import numpy as np
```

```python
api = req.get("https://api.blackiq.ir/sleep/sleep.json")
```

```python
sleep = api.json()
```

```python
amirmahdi = []

for i in range(len(sleep["nights"])):
    amirmahdi.append(sleep["nights"][i]["amirmahdi"])
```

```python
amirhossein = []

for i in range(len(sleep["nights"])):
    amirhossein.append(sleep["nights"][i]["amirhossein"])
```

```python
amirm = np.array(amirmahdi)
amirh = np.array(amirhossein)

plt.plot(amirm)
plt.plot(amirh)

plt.show()
```
